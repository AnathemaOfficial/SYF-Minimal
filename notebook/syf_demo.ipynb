# SYF-Minimal Demo

Non-canonical demonstration of the SYF invariant.

```
R = (F × E) / K
```

````python
import sys
sys.path.insert(0, '../python')
from syf_core import syf_core
````

## Test Vectors

````python
# Case 1: F=10, E=1.0, K=2 → R=5.0
r1 = syf_core(F=10, E=1.0, K=2)
print(f"Case 1: R = {r1}")
assert r1 == 5.0
````

````python
# Case 2: F=0, E=0.5, K=1 → R=0.0
r2 = syf_core(F=0, E=0.5, K=1)
print(f"Case 2: R = {r2}")
assert r2 == 0.0
````

````python
# Case 3: F=4, E=0.25, K=2 → R=0.5
r3 = syf_core(F=4, E=0.25, K=2)
print(f"Case 3: R = {r3}")
assert r3 == 0.5
````

## Invalid Inputs

````python
# F < 0 → Error
try:
    syf_core(F=-1, E=0.5, K=1)
except ValueError as e:
    print(f"Expected error: {e}")
````

````python
# E > 1 → Error
try:
    syf_core(F=1, E=1.5, K=1)
except ValueError as e:
    print(f"Expected error: {e}")
````

````python
# K <= 0 → Error
try:
    syf_core(F=1, E=0.5, K=0)
except ValueError as e:
    print(f"Expected error: {e}")
````
